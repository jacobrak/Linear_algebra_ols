```python
import numpy as np

def simple_ols(x, y):
    beta1 = np.sum((x - x.mean())*(y - y.mean())) / np.sum((x - x.mean())**2)
    beta0 = y.mean() - beta1 * x.mean()
    y_hat = beta0 +beta1*x
    return beta0, beta1 ,y_hat
```

```python
x = [1, 2, 4, 5, 12, 10,11]
y = [3, 2, 6, 8, 10, 14, 26]
```

```python
import matplotlib.pyplot as plt
beta0, beta1 ,y_hat = simple_ols(np.array(x), np.array(y))
x_line = np.linspace(0, np.array(x).max(), 15)  
y_line = beta0 + beta1 * x_line              

plt.scatter(x, y, label="Data")
plt.plot(y_line, color="purple", label="Regression Line")
plt.legend()
```
